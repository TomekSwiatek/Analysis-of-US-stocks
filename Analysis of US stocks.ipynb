## Indicators

```python
#Book Value
for i in range(1,df.Date.count()):
    df.loc[0,'Change in book value %'] = 100
    df.loc[i,'Change in book value %'] = round(((df.loc[i,'Book Value']- \
                                         df.loc[i-1,'Book Value']) \
                                         /df.loc[i-1,'Book Value'])*100,0)

#Increase in sales revenue
for i in range(1,df.Date.count()):
    df.loc[0,'Change in Revenue %'] = 100
    df.loc[i,'Change in Revenue %'] = round(((df.loc[i,'Revenue USD Mil']- \
                                             df.loc[i-1,'Revenue USD Mil']) \
                                             /df.loc[i-1,'Revenue USD Mil'])*100,0)

#Increase 
for i in range(1,df.Date.count()):
    df.loc[0,'Change in Net Income %'] = 100
    df.loc[i,'Change in Net Income %'] = round(((df.loc[i,'Net Income USD Mil']- \
                                                 df.loc[i-1,'Net Income USD Mil'])\
                                                 /df.loc[i-1,'Net Income USD Mil'])*100,0)
                                                      
```

```python
#Profitability
df['ROE %'] = df['Return on Equity %'] #best >20%
df['ROA %'] = df['Return on Assets %'] #best >10%
df['ROS %'] = df['Gross Margin %'] #best >10%
df['OM %'] = df['Operating Margin %'] #best >10%
df['NM %'] = df['Net Margin %'] #best >10%
```

```python
#Liquidity
df['Current Ratio'] = df['Current Ratio'] #best >2
df['Cash Ratio %'] = round((df['Cash & Short-Term Investments %']/df['Short-Term Debt %'])*100,2) #best >20%
df['Working capital'] = round((df['Total Current Assets %']-df['Short-Term Debt %']),0) #best >0
```

```python
#Activity
df['DSO'] = df['Days Sales Outstanding']
df['DI'] = df['Days Inventory']
df['PP'] = df['Payables Period']
df['Cash conversion cycle'] = round((df['DSO']+df['DI']-df['PP']),0) #best <0
```

```python
#Debt
df['Longterm Debt/Equity Ratio'] = df['Debt/Equity'] #best <20%
df['Debt ratio (Total debt/assets)'] = round(((df['Long-Term Debt %'] + df['Short-Term Debt %']) \
                                            /df['Total Assets %'])*100,2) #best <67%
```

```python
#Property
df['1st degree of coverage'] = round((df["Total Stockholders' Equity %"]/\
                                      (df['Net PP&E %']+df['Intangibles %']+df['Other Long-Term Assets %'])),2) #best >0.6
```

```python
#Cash flow
def flow(data):
        if data > 0: return 1
        else: return 0
        
df['Operating Cash Flow rate'] = 0        
df['Operating Cash Flow rate'] = df['Operating Cash Flow USD Mil'].apply(flow) #best >0

df['Investing Cash Flow rate'] = 0        
df['Investing Cash Flow rate'] = df['Cap Spending USD Mil'].apply(flow) #best <0

df['Finance Cash Flow rate'] = 0        
df['Finance Cash Flow rate'] = df['Cap Spending USD Mil'].apply(flow) #best <0

df['Free cash flow rate'] = 0        
df['Free cash flow rate'] = df['Free Cash Flow USD Mil'].apply(flow) #best >0
```

```python
#Cash flow description

def flows(data):
        if ((data['Operating Cash Flow rate'] > 0) & \
        (data['Investing Cash Flow rate'] > 0) & \
        (data['Finance Cash Flow rate'] > 0)) : return 'I - (+++) rare situation, need to analyze cases'      
        elif ((data['Operating Cash Flow rate'] > 0) & \
        (data['Investing Cash Flow rate']) > 0 & \
        (data['Finance Cash Flow rate']) == 0): return 'II - (++-) restructuring process'    
        elif ((data['Operating Cash Flow rate'] > 0) & \
        (data['Investing Cash Flow rate']) == 0 & \
        (data['Finance Cash Flow rate']) == 0): return 'III - (+--) stable company, the best'    
        elif ((data['Operating Cash Flow rate'] == 0) & \
        (data['Investing Cash Flow rate']) == 0 & \
        (data['Finance Cash Flow rate']) == 0) : return 'IV - (---) bankrupt'    
        elif ((data['Operating Cash Flow rate'] < 0) & \
        (data['Investing Cash Flow rate']) > 0 & \
        (data['Finance Cash Flow rate']) > 0)  : return 'V - (-++) temporary problems'    
        elif ((data['Operating Cash Flow rate'] == 0) & \
        (data['Investing Cash Flow rate']) == 0 & \
        (data['Finance Cash Flow rate']) > 0)  : return 'VI - (--+) start-up, early stage of activity'    
        elif ((data['Operating Cash Flow rate'] == 0) & \
        (data['Investing Cash Flow rate']) > 0 & \
        (data['Finance Cash Flow rate']) == 0)   : return 'VII - (-+-) probable bankruptcy'    
        elif ((data['Operating Cash Flow rate'] > 0) & \
        (data['Investing Cash Flow rate']) == 0 & \
        (data['Finance Cash Flow rate']) == 0)  : return 'VIII - (+-+) after the start-up phase'                                 
        else: return 'Error'

df['Cash flow description'] = '0'      
df['Cash flow description'] = df.apply(flows, axis=1)
```

## The rating of the company

```python
#Score based on fundamental indicators

def scores(data):
    l=0
    
    #Dyvidend
    if (data['Year-end Yield %'] > 2): l+=1;
    if (data['Dividends USD'] > 1): l+=1;
        
    #Book value
    if (data['Change in book value %'] > 0): l+=1;
    #Increase in sales revenue
    if (data['Change in Revenue %'] > 0): l+=1;
    #Increase in net income
    if (data['Change in Net Income %'] > 0): l+=1;
        
    #Profitability
    if (data['ROE %'] > 20): l+=1;
    if (data['ROA %'] > 10): l+=1;
    if (data['ROS %'] > 10): l+=1;
    if (data['OM %'] > 10): l+=1;
    if (data['NM %'] > 10): l+=1;
        
    #Liquidity   
    if (data['Current Ratio'] > 2): l+=1;
    if (data['Cash Ratio %'] > 20): l+=1;
    if (data['Working capital'] > 0): l+=1;
        
    #Activity    
    if (data['Cash conversion cycle'] < 0): l+=1;
        
    #Debt  
    if (data['Longterm Debt/Equity Ratio'] < 20): l+=1;
    if (data['Debt ratio (Total debt/assets)'] < 67): l+=1;
        
    #Cash flow description
    if (data['Cash flow description'] == 'III - (+--) stable company, the best'): l+=1    
    elif (data['Cash flow description'] == 'VI - (--+) start-up, early stage of activity'): l+=1
    elif (data['Cash flow description'] == 'VIII - (+-+) after the start-up phase'): l+=1;
        
    #Property
    if (data['1st degree of coverage'] > 0.6): l+=1;
        
    return l
        
df['Scores'] = '0'      
df['Scores'] = df.apply(scores, axis=1)            
```

Rating:
- 15-18: Very good company
- 12-14: Good company
- 9-11: Average company
- 0-9: Weak company

```python
#Final evaluation

def rate(score):
    if (score >= 15): return ('Very good company')
    elif (score >= 12): return ('Good company')
    elif (score >= 9): return ('Average company')
    else: return ('Weak company')

df['Rating'] = df['Scores'].apply(rate)
```
